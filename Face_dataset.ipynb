## Face Dataset 

```python
import os
import cv2
cam=cv2.VideoCapture(0)
cam.set(3, 640) # set video width
cam.set(4, 480) 
#face_detector=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
face_detector = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
if face_detector.empty():
    print("Error: Haar Cascade not loaded properly!")
else:
    print("loaded properly")
```

```python
# For each person, enter one numeric face id
name = input('\n enter user name')
print("\n Starting Camera to Take Photo's")
# Initialize individual sampling face count
count = 0
if not os.path.exists("dataset"):
    os.makedirs("dataset")
while(True):
    ret, img = cam.read()
    if not ret: # if camera not capture frame then exit
        break
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_detector.detectMultiScale(gray, 1.3, 5)
    print("Press 'y' to save image, 'q' to quit")
    for (x,y,w,h) in faces:
        cv2.rectangle(img, (x,y), (x+w,y+h), (255,0,0), 2)     
    cv2.imshow("image", img)
    key=cv2.waitKey(1) & 0xFF
    if key == ord('y') and len(faces) > 0:
        for (x, y, w, h) in faces:
            count += 1
            cv2.imwrite(f"dataset/User.{name}.{count}.jpg", gray[y:y+h, x:x+w])
            print(f"Saved image {count}")
        if count >= 10:  # limit to 10 images
            break
    elif key == ord('q'):
        break
cam.release()
cv2.destroyAllWindows()
```
